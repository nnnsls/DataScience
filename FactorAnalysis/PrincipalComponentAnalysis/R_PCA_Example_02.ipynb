```r
library("car")
csv = read.csv('../b10.csv')
head(csv[,2:10])
```

```r
fit=prcomp(csv[,2:10], scale=TRUE)
fit
summary(fit)
```

```r
par(mfrow=c(2,2))
plot(fit$x[,1],fit$x[,2],col=0)
text(fit$x[,1],fit$x[,2],csv[,11])
plot(fit$x[,1],fit$x[,3],col=0)
text(fit$x[,1],fit$x[,3],csv[,11])
plot(fit$x[,3],fit$x[,2],col=0)
text(fit$x[,3],fit$x[,2],csv[,11])
plot(fit$x[,2],fit$x[,4],col=0)
text(fit$x[,2],fit$x[,4],csv[,11])
```

```r
par(mfrow=c(2,2))
plot(fit$x[,1],fit$x[,2],col=0)
text(fit$x[,1],fit$x[,2],csv[,12])
plot(fit$x[,1],fit$x[,3],col=0)
text(fit$x[,1],fit$x[,3],csv[,12])
plot(fit$x[,3],fit$x[,2],col=0)
text(fit$x[,3],fit$x[,2],csv[,12])
plot(fit$x[,2],fit$x[,4],col=0)
text(fit$x[,2],fit$x[,4],csv[,12])
```
