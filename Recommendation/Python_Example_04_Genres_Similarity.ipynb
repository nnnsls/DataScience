## Cosine Similarity
Cosine similarity is a useful function in machine learning, let us classify or cluster easily. However, the genres similarity is not useful in MovieLens Datasets, you can try others data from IMDb.com!

```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
csv_movies = pd.read_csv('Data/ml-latest-small/movies.csv')
```

```python
csv_movies['Action'] = csv_movies.apply(lambda x: 'Action' in x['genres'], axis=1)+0
csv_movies['Adventure'] = csv_movies.apply(lambda x: 'Adventure' in x['genres'], axis=1)+0
csv_movies['Animation'] = csv_movies.apply(lambda x: 'Animation' in x['genres'], axis=1)+0
csv_movies['Children'] = csv_movies.apply(lambda x: 'Children' in x['genres'], axis=1)+0
csv_movies['Comedy'] = csv_movies.apply(lambda x: 'Comedy' in x['genres'], axis=1)+0
csv_movies['Crime'] = csv_movies.apply(lambda x: 'Crime' in x['genres'], axis=1)+0
csv_movies['Documentary'] = csv_movies.apply(lambda x: 'Documentary' in x['genres'], axis=1)+0
csv_movies['Drama'] = csv_movies.apply(lambda x: 'Drama' in x['genres'], axis=1)+0
csv_movies['Fantasy'] = csv_movies.apply(lambda x: 'Fantasy' in x['genres'], axis=1)+0
csv_movies['Film-Noir'] = csv_movies.apply(lambda x: 'Film-Noir' in x['genres'], axis=1)+0
csv_movies['Horror'] = csv_movies.apply(lambda x: 'Horror' in x['genres'], axis=1)+0
csv_movies['Musical'] = csv_movies.apply(lambda x: 'Musical' in x['genres'], axis=1)+0
csv_movies['Mystery'] = csv_movies.apply(lambda x: 'Mystery' in x['genres'], axis=1)+0
csv_movies['Romance'] = csv_movies.apply(lambda x: 'Romance' in x['genres'], axis=1)+0
csv_movies['Sci-Fi'] = csv_movies.apply(lambda x: 'Sci-Fi' in x['genres'], axis=1)+0
csv_movies['Thriller'] = csv_movies.apply(lambda x: 'Thriller' in x['genres'], axis=1)+0
csv_movies['War'] = csv_movies.apply(lambda x: 'War' in x['genres'], axis=1)+0
csv_movies['Western'] = csv_movies.apply(lambda x: 'Western' in x['genres'], axis=1)+0
```

```python
movies_genres = csv_movies.set_index('movieId').loc[:,'Action':]
movies_genres.head()
```

```python
genres_similarity = cosine_similarity(csv_movies.loc[:,'Action':])
pd_genres_similarity = pd.DataFrame(genres_similarity, index=csv_movies.movieId, columns=csv_movies.movieId)
```
