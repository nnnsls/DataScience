```python
import pandas as pd
import urllib
from bs4 import BeautifulSoup
```

```python
csv_links = pd.read_csv('Data/ml-latest-small/links.csv')[0:10]
```

```python
csv_links
```

```python
csv_links['Synopsis'] = None
fail_list = []

for index, row in csv_links.iterrows():
    if row['Synopsis'] is None:
        try:
            synopsis_url = 'http://www.imdb.com/title/tt%s/synopsis?ref_=tt_stry_pl' % row['imdbId']

            response = urllib.request.urlopen(synopsis_url)
            soup = BeautifulSoup(response, "lxml")
            csv_links.loc[index,'Synopsis'] = soup.find("div", class_='display').text.strip()
            print('Completed:', row['movieId'], ' | ', csv_links.loc[index,'Synopsis'][0:30])
        except:
            fail_list.append([row['movieId'], row['imdbId']])
            print('Failed:', row['movieId'], ' | URL:', row['imdbId'])
```
