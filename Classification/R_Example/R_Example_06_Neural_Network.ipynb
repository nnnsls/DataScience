```r
library(nnet)
set.seed(852)
data = read.csv('../ACollege1995R.csv')
```

```r
stand<-function(x) {
   n<-dim(x)[1]
   p<-dim(x)[2]
   m<-apply(x,2,mean)
   s<-apply(x,2,sd)
   m<-matrix(m,nr=n,nc=p,byrow=T)
   s<-matrix(s,nr=n,nc=p,byrow=T)
   (x-m)/s
}
```

```r
ann = function(x,y,size,maxit=100,linout=F,try=5) {
  ann1 = nnet(y~.,data=x,size=size,maxit=maxit,linout=linout)
  v1 = ann1$value
  for (i in 2:try) {
    ann = nnet(y~.,data=x,size=size,maxit=maxit,linout=linout)
    if (ann$value<v1) {
      v1 = ann$value
      ann1 = ann
    }
  }
  ann1
}
```

```r
geterror = function(model,y1,d2,y2) {
  pred = round(0+(model$fitted.values>0.5))
  tt = table(pred,y1)
  error0 = (tt[2]+tt[3])/sum(tt)
  pred = 0+(predict(model,d2)>0.5)
  tt = table(pred,y2)
  error1 = (tt[2]+tt[3])/sum(tt)
  cat('Training error:',error0,'| Testing error:',error1,'\n')
}
```

```r
id = sample(1:dim(data)[1], size=dim(data)[1]*0.7,replace=F)
train = data[id,]
test = data[-id,]
```

```r
ann8 = ann(train[,1:13],train$good,8,linout=F,try=30)
```

```r
ann9 = ann(train[,1:13],train$good,9,linout=F,try=30)
```

```r
ann10 = ann(train[,1:13],train$good,10,linout=F,try=30)
```

```r
geterror(ann8,train$good,test[,1:13],test$good)
geterror(ann9,train$good,test[,1:13],test$good)
geterror(ann10,train$good,test[,1:13],test$good)
```
